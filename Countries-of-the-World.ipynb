* Currency_Code: Currency code used in the country.
* Fertility Rate: Average number of children born to a woman during her lifetime.
* Forested Area (%): Percentage of land area covered by forests.
* Gasoline_Price: Price of gasoline per liter in local currency.
* GDP: Gross Domestic Product, the total value of goods and services produced in the country.
* Gross Primary Education Enrollment (%): Gross enrollment ratio for primary education.
* Gross Tertiary Education Enrollment (%): Gross enrollment ratio for tertiary education.
* Infant Mortality: Number of deaths per 1,000 live births before reaching one year of age.
* Largest City: Name of the country's largest city.
* Life Expectancy: Average number of years a newborn is expected to live.
* Maternal Mortality Ratio: Number of maternal deaths per 100,000 live births.
* Minimum Wage: Minimum wage level in local currency.
* Official Language: Official language(s) spoken in the country.
* Out of Pocket Health Expenditure (%): Percentage of total health expenditure paid out-of-pocket by individuals.
* Physicians per Thousand: Number of physicians per thousand people.
* Population: Total population of the country.
* Population: Labor Force Participation (%): Percentage of the population that is part of the labor force.
* Tax Revenue (%): Tax revenue as a percentage of GDP.
* Total Tax Rate: Overall tax burden as a percentage of commercial profits.
* Unemployment Rate: Percentage of the labor force that is unemployed.
* Urban Population: Percentage of the population living in urban areas.
* Latitude: Latitude coordinate of the country's location.
* Longitude: Longitude coordinate of the country's location.

## Exploration of Pair-Wise Regression Analyses.

```python
import statsmodels.api as sm

def regression_f2(data, variable1, variable2):
    
    # Step 1: Prepare your data
    dataxna = data[[variable1, variable2]].dropna()
    X = dataxna[[variable1]]  # Predictor variable(s)
    X = sm.add_constant(X)   # Add a constant term for the intercept
    y = dataxna[variable2]    # Target variable

    # Step 2: Fit the model with your data
    model = sm.OLS(y, X).fit()

    # Step 3: Access the results and statistics
    coefficients = model.params
    summary = model.summary()

    # Output the results
    print("Coefficients:")
    print(coefficients)
    print("\nSummary:")
    print(summary)
```

```python
regression_f2(wd23, 'GDP', 'Gasoline Price')
```

In a linear regression analysis, the results often include several figures and statistics that provide insights into the model's performance, the significance of the predictor variables, and the overall fit of the model. Here are some important figures and statistics commonly found in regression results and their interpretations:

1. Coefficients: The coefficients (β₁, β₂, etc.) represent the estimated effects of the predictor variables on the target variable. They indicate the change in the target variable's value associated with a one-unit change in the corresponding predictor variable while holding other variables constant. Positive coefficients suggest a positive relationship, and negative coefficients suggest a negative relationship.

2. Intercept (β₀): The intercept represents the expected mean value of the target variable when all predictor variables are zero. It is the baseline value of the target variable.

3. Standard Errors: Standard errors measure the precision of the coefficient estimates. Smaller standard errors indicate more precise estimates. They are used to calculate confidence intervals and conduct hypothesis tests.

4. t-statistics: The t-statistics assess the significance of each coefficient. It measures the ratio of the coefficient estimate to its standard error. Larger absolute t-values indicate more significant predictor variables. The associated p-values indicate the probability of observing the coefficient value if the null hypothesis (no effect) were true. Lower p-values (< 0.05) suggest significant predictor variables.

5. R-squared (R²): R-squared is a measure of how well the regression model fits the data. It represents the proportion of the variance in the target variable that is explained by the predictor variables. R-squared values range from 0 to 1, with higher values indicating a better fit. However, R-squared alone should not be the sole criterion for model evaluation.

6. Adjusted R-squared: Adjusted R-squared adjusts the R-squared value for the number of predictor variables and the sample size. It penalizes the addition of unnecessary predictors. It provides a more conservative estimate of the model's fit and is particularly useful when comparing models with different numbers of predictors.

7. F-statistic: The F-statistic tests the overall significance of the regression model. It assesses whether the overall relationship between the predictors and the target variable is significant. A large F-statistic with a small associated p-value indicates a significant relationship.

8. Durbin-Watson statistic: The Durbin-Watson statistic measures the presence of autocorrelation (correlation of error terms) in the model. It ranges from 0 to 4, with values around 2 suggesting no autocorrelation.

9. Residuals: Residuals are the differences between the observed values and the predicted values by the regression model. Examining the residuals can help assess the model's assumptions, such as linearity, homoscedasticity (constant variance), and normality.

These figures and statistics provide insights into the model's significance, goodness-of-fit, and the individual contributions of the predictor variables. It is essential to carefully interpret and consider these results in the context of your specific analysis and research question.

```python
import statsmodels.api as sm

# Check if the DataFrame exists
if 'reg_pair_stat' in locals():
    # Delete the DataFrame
    del reg_pair_stat

# Create a table to be populated with some important summary stats for each pairing
reg_pair_stat = pd.DataFrame(columns=['reg_pair', 'pred_coef', 'p-value', 'std_err', 'r^2', 'adjr^2'])

def regression_f3(data, variable1, variable2):
    
    # Declare reg_pair_stat as a global variable
    global reg_pair_stat
    
    # Step 1: Prepare your data
    dataxna = data[[variable1, variable2]].dropna()
    X = dataxna[[variable1]]  # Predictor variable(s)
    X = sm.add_constant(X)   # Add a constant term for the intercept
    y = dataxna[variable2]    # Target variable

    # Step 2: Fit the model with your data
    model = sm.OLS(y, X).fit()

    # Step 3: Access the results and statistics
    coefficients = model.params
    r_squared = model.rsquared
    adj_r_squared = model.rsquared_adj
    summary = model.summary()

    # Extract specific information from the summary
    predictor_coef = float(summary.tables[1].data[2][1])
    std_err = float(summary.tables[1].data[2][2])
    p_value = float(summary.tables[1].data[2][4])

    # Populate 'reg_pair_stat' with chosen summary stats
    reg_pair_stat = pd.concat([reg_pair_stat, pd.DataFrame({'reg_pair': f"{variable1} predicting {variable2}", 
                                                            'pred_coef': predictor_coef, 
                                                            'p-value': p_value, 
                                                            'std_err': std_err, 
                                                            'r^2': r_squared, 
                                                            'adjr^2': adj_r_squared}, index=[0])], ignore_index=True)
```

```python
# Perform regression analysis for all pairs of numerical columns
for pred in numerical_columns:
    for target in numerical_columns:
        if pred != target:  # Skip if predictor and target are the same column
            regression_f3(wd23, pred, target)
```

## PCA Analysis

Processing the data for PCA.

```python
# Remove irrelevant columns
columns_to_remove = ['Country', 'Abbreviation', 'Capital/Major City', 'Currency-Code', 'Official language', 'Largest city']
wd23_processed = wd23.drop(columns_to_remove, axis=1)

# Impute missing values using KNNImputer from scikit-learn
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=3)
wd23_processed = pd.DataFrame(imputer.fit_transform(wd23_processed), columns=wd23_processed.columns)

wd23_processed
```

Scaling the data for PCA.

```python
from sklearn.preprocessing import StandardScaler

# Assuming 'wd23_processed' is the preprocessed DataFrame

# Perform standardization
scaler = StandardScaler()
wd23_scaled = pd.DataFrame(scaler.fit_transform(wd23_processed), columns=wd23_processed.columns)
```

Performing PCA on the scaled data frame.

```python
from sklearn.decomposition import PCA
number_components = 7

# Assuming 'wd23_scaled' is the standardized DataFrame

# Perform PCA
pca = PCA(n_components=number_components)  # Specify the number of components
principal_components = pca.fit_transform(wd23_scaled)

# The principal components are now obtained and stored in the 'principal_components' array or DataFrame
```

```python
import matplotlib.pyplot as plt

# Assuming 'principal_components' is the array or DataFrame of the principal components obtained from PCA

# Step 1: Analyze the explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio:")
print(explained_variance_ratio)

# Step 2: Evaluate cumulative explained variance
cumulative_variance_ratio = explained_variance_ratio.cumsum()
print("Cumulative Explained Variance:")
print(cumulative_variance_ratio)

# Step 3: Examine loadings
loadings = pd.DataFrame(pca.components_.T, columns=[f"PC{i}" for i in range(1, number_components+1)], index=wd23_processed.columns)
print("Loadings:")
print(loadings)

# Step 4: Optional visualization
# Scatter plot of the first two principal components
plt.scatter(principal_components[:, 0], principal_components[:, 1])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Scatter Plot of PC1 vs. PC2")
plt.show()
```

```python
index_mapping = {
'Population: Labor force participation (%)': 'LF participation (%)',
'Gross tertiary education enrollment (%)': 'GTH enrollment (%)',
'Gross primary education enrollment (%)': 'GPE enrollment (%)',
'Out of pocket health expenditure': 'OoP health expense',
'Density\n(P/Km2)': 'Density (P/Km2)'}
loadings = loadings.rename(index=index_mapping)
loadings = loadings.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
import matplotlib.pyplot as plt

def visualize_loadings(loadings, component_number):
    
    # Get the loadings for the specified component
    component_loadings = loadings[f"PC{component_number}"]
        
    # Sort the loadings in descending order
    sorted_loadings = component_loadings.sort_values(ascending=False)

    # Create a bar plot to visualize the loadings
    plt.figure(figsize=(40, 24))
    plt.bar(sorted_loadings.index, sorted_loadings.values)
    plt.xticks(rotation=45, ha='right')
    plt.xlabel("Variable")
    plt.ylabel("Loading")
    plt.title(f"Loadings for PC{component_number}")
    
    plt.xticks(range(len(sorted_loadings.index)), sorted_loadings.index)  # Force all ticks to be displayed
    
    plt.tight_layout(pad=15)
    plt.rcParams.update({'font.size': 20})
    plt.show()
```

```python
PCs = range(1, number_components+1)

for PC in PCs:
    visualize_loadings(loadings, PC)
```
